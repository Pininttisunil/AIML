## part-1:Lists / Arrays / type


```python
import numpy as np
import pandas as pd

list1=[1,2,3]
list2=[4,5,6]

arr1=np.array(list1)
arr2=np.array(list2)

print(type(list1))

print(arr1)
print(arr2)
```

 **Mathematical operations**
 **Multipication**

```python
mul = arr1*arr2
print("array1 multiply by array2:",mul)
```

**division**

```python
div = arr1/arr2
print("array1 divided by array2:",div)
```

**Addition**

```python
add = arr1+arr2
print("array1 addition with array2:",add)
```

**Poweroff**

```python
print(f"{arr2} rasied to the power of {arr1}:{arr2**arr1}")
```

# Apply sine fuctions



```python
#sine function
print("sine:",np.sin(arr1))
#log functions
print("natural log:",np.log(arr1))
print("base-10 log:",np.log10(arr1))
print("base-2 log:",np.log2(arr1))
#expontial
print("exponential:",np.exp(arr1))
```

# **Part-2:Visualization of Data using matplotlib, pyplots Packages**

# **SCATTER PLOT**

```python
people = ['kiran', 'arun','sunil','vijay', 'varun']
age =[25, 30, 35, 40, 45]
height =[145, 151, 165, 173,345]
weight=[45, 55, 65, 75,54]
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(weight,height)
plt.title("Relationship between the Weight and Height of paients")
plt.ylabel("height in meters")
plt.xlabel("Weights of the paients in kgs")
plt.show()
```

# **AGE TO WEIGHT**

```python
plt.scatter(age,weight)
plt.title("Relationship between the age and weights of paients")
plt.ylabel("weights in meters")
plt.xlabel("ages of the paients")
plt.show()
```

# **AGE TO HEIGHT**

```python
plt.scatter(age,height)
plt.title("Relationship between the age and Height of paients")
plt.ylabel("height in meters")
plt.xlabel("Ages of the paients")
plt.show()
```

# **GENERATE A BAR CHAT**

```python
plt.figure(figsize=(5,4))
plt.title("people's weight in kgs",fontsize=20,fontstyle='italic')

#main plot function 'bar'

plt.bar(x=people,height=weight,width=0.6,color='red',edgecolor='k',alpha=0.9)
plt.xlabel("people",fontsize=16)
plt.xticks(fontsize=14,rotation=30)
plt.yticks(fontsize=14)
plt.ylabel("weight (in kgs)",fontsize=16)
plt.show()
```

# **HISTOGRAM**

```python
import numpy as np
plt.figure(figsize=(7,5))

#main plot
plt.hist(weight,color='red',edgecolor='k',alpha=0.6,bins=5)

plt.title("Histogram of paients weight",fontsize=16)
plt.xlabel("weight in kgs",fontsize=13)

plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
